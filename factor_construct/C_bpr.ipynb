# bm
Book-to-market ratio, which equals the book value of equity divided by market capitalization

```python
# 04财务指标分析/FI_T10
# F101001A
df = pd.read_csv('../../Data/raw_data/04财务指标分析/FI_T10.csv')
df = df[df['Stkcd'].map(lambda x: (x>=0 and x<=9999) or (x>=600000 and x<=609999))]
df['Accper'] = pd.DatetimeIndex(df['Accper'])
df
```

```python
bm = df.pivot_table(values='F101001A',index='Accper',columns='Stkcd')
bm
```

```python
    df2 = pd.read_csv('../../Data/raw_data/18年、中、季报基本情况/IAR_Rept.csv')
    df2 = df2[df2['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<= 609999))]
    df2['Accper'] = pd.DatetimeIndex(df2['Accper'])
    df2['Annodt'] = pd.DatetimeIndex(df2['Annodt'])
    df2
```

```python
#  将报表会计截止日替换为报表公布日
def change_date(df1, value):
    '''
    df1: old dates that values have,
    df2: old dates mapped to new dates,
    value: values need to be reindexed
    '''
    df2 = pd.read_csv('../../Data/raw_data/18年、中、季报基本情况/IAR_Rept.csv')
    df2 = df2[df2['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<= 609999))]
    df2['Accper'] = pd.DatetimeIndex(df2['Accper'])
    df2['Annodt'] = pd.DatetimeIndex(df2['Annodt'])

    date_change = pd.merge(df1[['Stkcd','Accper']], df2[['Stkcd','Accper','Annodt']], how='inner', on=['Stkcd','Accper'])
    date_change = date_change.sort_values(by='Annodt')
    new_df = pd.DataFrame()
    for index in tqdm(date_change.index):  # 为统一用.loc取元素
        new_df.loc[date_change.loc[index, 'Annodt'], date_change.loc[index, 'Stkcd']] = \
        value.loc[date_change.loc[index, 'Accper'], date_change.loc[index, 'Stkcd']]
    return new_df

bm2 = change_date(df[['Stkcd','Accper']],bm)
bm2
```

```python
bm2.to_csv('../../Data/factors/bm.csv')
```

# cfp
Operating cash flows divided by quarter-end market capitalization. Related data are
contained in quarterly reports and acquired from CSMAR.

```python
# month-end market capitalization
msmvttl = mm.extract_variable_oneFile(filePath='../../Data/raw_data/02月个股回报率/TRD_Mnth.csv',
                                     fileType='csv',
                                     variable='Msmvttl',
                                     time_col='Trdmnt',
                                     code_col='Stkcd')
msmvttl.index = msmvttl.index.to_period('M').to_timestamp('M')
msmvttl
```

```python
# Operating cash flows
# C001001000 [销售商品、提供劳务收到的现金]
df = pd.read_csv('../../Data/raw_data/15现金流量表（直接法）/FS_Comscfd.csv')
df = df[df['Typrep'] == 'A']
df = df[df['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<=609999))]
df['Accper'] = pd.DatetimeIndex(df['Accper'])
df
```

```python
oprtcf = df.pivot_table(values='C001001000', index='Accper', columns='Stkcd')
oprtcf
```

```python
# 解决'C001001000'中没有当日数据的问题，但其他并列指标有，故df中有。虽然只缺了1997-06-30中一天
accper_date = df[['Stkcd','Accper']]
accper_date = accper_date[accper_date['Accper'].map(lambda x: x in oprtcf.index)]
accper_date
```

```python
#  将报表会计截止日替换为报表公布日
def change_date(df1, value):
    '''
    df1: old dates that values have,
    df2: old dates mapped to new dates,
    value: values need to be reindexed
    '''
    df2 = pd.read_csv('../../Data/raw_data/18年、中、季报基本情况/IAR_Rept.csv')
    df2 = df2[df2['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<= 609999))]
    df2['Accper'] = pd.DatetimeIndex(df2['Accper'])
    df2['Annodt'] = pd.DatetimeIndex(df2['Annodt'])

    date_change = pd.merge(df1[['Stkcd','Accper']], df2[['Stkcd','Accper','Annodt']], how='inner', on=['Stkcd','Accper'])
    date_change = date_change.sort_values(by='Annodt')
    value = value.reindex(date_change['Stkcd'].drop_duplicates(),axis=1)
    new_df = pd.DataFrame()
    for index in tqdm(date_change.index):  # 为统一用.loc取元素
        new_df.loc[date_change.loc[index, 'Annodt'], date_change.loc[index, 'Stkcd']] = \
        value.loc[date_change.loc[index, 'Accper'], date_change.loc[index, 'Stkcd']]
    return new_df

oprtcf2 = change_date(accper_date,oprtcf)
oprtcf2
```

```python
# 日期对齐
oprtcf2 = oprtcf2.append(pd.DataFrame(index=msmvttl.index)).sort_index()
oprtcf2 = oprtcf2.fillna(method='ffill')
oprtcf2
```

```python
cfp = oprtcf2 / msmvttl
cfp = cfp.dropna(how='all')
cfp
```

```python
cfp.to_csv('../../Data/factors/cfp.csv')
```

# cashspr
Cash productivity, which is defined as quarter-end market capitalization plus long-term debt minus total assets divided by cash and equivalents. Related data are contained in
quarterly reports and acquired from CSMAR.  
(quarter-end market capitalization + long-term debt - total assets) / cash and equivalents

## 1. market capitalization

```python
msmvttl = mm.extract_variable_oneFile(filePath='../../Data/raw_data/02月个股回报率/TRD_Mnth.csv',
                                     fileType='csv',
                                     variable='Msmvttl',
                                     time_col='Trdmnt',
                                     code_col='Stkcd')
msmvttl.index = msmvttl.index.to_period('M').to_timestamp('M')
msmvttl
```

## 2. long-term debt

```python
# A002206000 [长期负债合计] - 长期负债各项目之合计。1990年起使用，2007年停止使用。
# A002200000 [非流动负债合计] - 所有非流动负债的合计。 2007年起使用。
# 二者拼接起来
# ???为什么这两项时间仍都一直使用
# 决定，只用A002206000 [长期负债合计]
df = pd.read_csv('../../Data/raw_data/16资产负债表/FS_Combas.csv')
df = df[df['Typrep'] == 'A']
df = df[df['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<=609999))]
df['Accper'] = pd.DatetimeIndex(df['Accper'])
df
```

```python
ltd = df.pivot_table(values='A002206000', index='Accper', columns='Stkcd')
ltd = ltd.replace(0,np.nan)
ltd = ltd.fillna(method='ffill')
ltd
```

## 3. total assets

```python
# A001000000 [资产总计] - 资产各项目之总计。1990年起使用
tta = df.pivot_table(values='A001000000', index='Accper', columns='Stkcd')
tta = tta.replace(0,np.nan)
tta = tta.fillna(method='ffill')
tta
```

## 4. cash and equivalents

```python
# C006000000 [期末现金及现金等价物余额] - 期末现金及现金等价物的余额.2007年开始使用。
df = pd.read_csv('../../Data/raw_data/15现金流量表（直接法）/FS_Comscfd.csv')
df = df[df['Typrep'] == 'A']
df = df[df['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<=609999))]
df['Accper'] = pd.DatetimeIndex(df['Accper'])
df
```

```python
cce = df.pivot_table(values='C006000000', index='Accper', columns='Stkcd')
cce = cce.replace(0,np.nan)
cce = cce.fillna(method='ffill')
cce
```

## 5. results

```python
cashspr = (msmvttl + ltd - tta) / cce
cashspr = cashspr.loc['1997-06-30':]  # cash and cash equivalent最早在1997-06-30
cashspr = cashspr.fillna(method='ffill')
cashspr
```

```python
cashspr.T.to_csv('../../Data/factors/cashspr.csv')
```

# invest
The sum of annual change in fixed assets and annual change in inventories divided by
lagged total assets.

## 1. annual change in fixed assets

```python
# A001212000 [固定资产净额] - 固定资产净额为固定资产原价除去累计折旧和固定资产减值准备之后的净额。
df = pd.read_csv('../../Data/raw_data/16资产负债表/FS_Combas.csv')
df = df[df['Typrep'] == 'A']
df = df[df['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<=609999))]
df['Accper'] = pd.DatetimeIndex(df['Accper'])
df
```

```python
fa = df.pivot_table(values='A001212000', index='Accper', columns='Stkcd')
fa = fa[fa.index.month == 12]
fa
```

```python
pct_fa = fa.pct_change(fill_method=None)
pct_fa
```

## 2. annual change in inventories divided by lagged total assets

```python
df = pd.read_csv('../../Data/raw_data/16资产负债表/FS_Combas.csv')
df = df[df['Typrep'] == 'A']
df = df[df['Stkcd'].map(lambda x: (x>=0 and x<=9999) or (x>=600000 and x<=609999))]
df['Accper'] = pd.DatetimeIndex(df['Accper'])
df
```

```python
# A001000000资产总计
tta = df.pivot_table(values='A001000000', index='Accper', columns='Stkcd')
tta = tta[tta.index.month == 12]
tta = tta.replace(0, np.nan)
tta
```

```python
# A001123000 [存货净额] - “存货”与“存货跌价准备”之差额。
inventory = df.pivot_table(values='A001123000', index='Accper', columns='Stkcd')
inventory = inventory[inventory.index.month == 12]
inventory = inventory.replace(0, np.nan)
inventory
```

```python
inv_by_tta = inventory / tta.shift(1)
inv_by_tta
```

```python
pct_inv_by_tta = inv_by_tta.pct_change(fill_method=None)
pct_inv_by_tta
```

## 3. sum

```python
invest = pct_fa + pct_inv_by_tta
monthEnd = pd.date_range(start='1990-12-31', end='2022-01-31', freq='m')
invest = invest.reindex(monthEnd)
invest = invest.fillna(method='ffill')
invest.to_csv('../../Data/factors/invest.csv')
```

# realestate
Investment real estates divided by fixed assets.

```python
df = pd.read_csv('../../Data/raw_data/16资产负债表/FS_Combas.csv')
df = df[df['Typrep'] == 'A']
df = df[df['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<=609999))]
df['Accper'] = pd.DatetimeIndex(df['Accper'])
# A001212000 [固定资产净额] - 固定资产净额为固定资产原价除去累计折旧和固定资产减值准备之后的净额。
fa = df.pivot_table(values='A001212000', index='Accper', columns='Stkcd')
fa = fa.fillna(method='ffill')
fa
```

# depr
Depreciation divided by fixed assets.

```python
# 这张表的csv文件也有问题，故改用excel
df1 = pd.read_excel('../../Data/raw_data/19固定资产、累计折旧、减值准备/FN_Fn020.xlsx')
df2 = pd.read_excel('../../Data/raw_data/19固定资产、累计折旧、减值准备/FN_Fn0201.xlsx')
df = pd.concat([df1, df2])
df
```

```python
df = df[df['Stkcd'].map(lambda x: x.startswith('00') or x.startswith('60'))]
df['Stkcd'] = df['Stkcd'].map(lambda x: int(x))
# Typrep [报表类型] - 1、合并会计报表；2、母公司会计报表
df = df[df['Typrep'] == 1]
# Fn02001 [科目类型] - 1、原值；2、累计折旧；3、净值；4、减值准备；5、净额
df = df[df['Fn02001'] == 2]
df['Accper'] = pd.DatetimeIndex(df['Accper'])
# Fn02013 [期末余额] - 固定资产的期末余额
df['Fn02013'] = df['Fn02013'].map(lambda x: float(x))
```

```python
depr = df.pivot_table(values='Fn02013', index='Accper', columns='Stkcd')
depr
```

```python
df = pd.read_csv('../../Data/raw_data/16资产负债表/FS_Combas.csv')
df = df[df['Typrep'] == 'A']
df = df[df['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<=609999))]
df['Accper'] = pd.DatetimeIndex(df['Accper'])
# A001212000 [固定资产净额] - 固定资产净额为固定资产原价除去累计折旧和固定资产减值准备之后的净额。
fa = df.pivot_table(values='A001212000', index='Accper', columns='Stkcd')
fa = fa.replace(0,np.nan)
fa
```

```python
depr_fa = depr / fa
depr_fa
```
