```python
import my_methods as mm
import pandas as pd
import numpy as np
import datetime as dt
from sklearn.linear_model import LinearRegression
"""
股票周度回报率关于市值加权市场回报率的beta
股票周度回报率，市值加权市场周度回报率
取样时长为当月的前三年，且至少有52周数据
"""
```

# beta

```python
df1 = pd.read_csv('../../Data/raw_data/12周个股回报率/TRD_Week.csv')
df2 = pd.read_csv('../../Data/raw_data/12周个股回报率/TRD_Week1.csv')
df3 = pd.read_csv('../../Data/raw_data/12周个股回报率/TRD_Week2.csv')
df = pd.concat([df1,df2,df3])
df = df[df['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<=609999))]
df['Trdwnt'] = pd.to_datetime([x+'-5' for x in df['Trdwnt']], format='%Y-%W-%w') - pd.DateOffset(weeks=1)
df
```

```python
wret = df.pivot_table(values='Wretwd', index='Trdwnt', columns='Stkcd')
wret
```

```python
df = pd.read_csv('../../Data/raw_data/13周市场回报率/TRD_Weekcm.csv', index_col='Trdwnt')
df
```

```python
# YY-WW格式转换为日期必须明确是周内的第几天，0为周日(最后一天)，5为周五
# 日期format: %W指当年第几周，%w指当周第几天
# [x... for x in y if x...]可快速转换序列
# 但CSMAR原始数据与pandas.to_datetime对当年第几周的理解似乎有些差异(观察最后一天2022/2/18)
# 故再减去7天的DatetimeDelta
df.index = pd.to_datetime([x+'-5' for x in df.index], format='%Y-%W-%w') - pd.DateOffset(days=7)
df
```

```python
market_wret = df[df['Markettype'] == 5].loc[:,'Cwretwdtl']  # 5=沪深A股市场（不包含科创板、创业板）
# Cwretwdtl [考虑现金红利再投资的综合周市场回报率(总市值加权平均法)]
market_wret
```

```python
wbeta = pd.DataFrame(columns=wret.columns, index=wret.index)
model = LinearRegression()
for endday in wret.index:
    print(endday)
    beginday = endday - pd.DateOffset(years=3)
    for stock in wret.columns:
        y = wret.loc[beginday:endday,stock]
        x = market_wret[beginday:endday]
        s = pd.concat([y,x], axis=1)
        s = s.dropna()
        if len(s) <52:
            continue
        y = s.iloc[:,0].to_numpy()
        X = s.iloc[:,[1]].to_numpy()
        model.fit(y=y, X=X)
        wbeta.loc[endday, stock] = model.coef_[0]
```

```python
wbeta
```

```python
wbeta.to_csv('../../Data/factors/beta.csv')
```

# betasq
