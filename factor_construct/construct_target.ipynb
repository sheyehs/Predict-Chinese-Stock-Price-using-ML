```python
import my_methods as mm
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df = pd.read_csv('../../Data/raw_data/02月个股回报率/TRD_Mnth.csv')
df = df[df['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<= 609999))]
df['Trdmnt'] = pd.DatetimeIndex(df['Trdmnt']).to_period(freq='M').to_timestamp(freq='M')
df
```

```python
monthly_return = df.pivot_table(values='Mretwd', index='Trdmnt', columns='Stkcd')
monthly_return
```

```python
future_1_monthly_return = monthly_return.shift(-1).dropna(how='all')
future_1_monthly_return
```

```python
future_1_monthly_return.to_csv('../../Data/future_1_monthly_return.csv')
```
