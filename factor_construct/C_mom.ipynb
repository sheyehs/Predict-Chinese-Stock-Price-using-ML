```python
import my_methods as mm
import pandas as pd
import numpy as np
import datetime as dt
```

# mom1m
1-month cumulative return.

```python
df = pd.read_csv('../../Data/raw_data/02月个股回报率/TRD_Mnth.csv')
df['Trdmnt'] = pd.DatetimeIndex(df['Trdmnt']).to_period('M').to_timestamp('M')  # 快速将月首转到月末
df
```

```python
mret = df.pivot_table(values='Mretwd',index='Trdmnt',columns='Stkcd')
mret
```

```python
mret.to_csv('../../Data/factors/mom1m.csv')
```

# mom6m
 5-month cumulative returns ending one month before month end.

```python
mret6_1 = mret.rolling(window=5, closed='left').apply(lambda x: np.prod(x+1)-1)
mret6_1
```

```python
mret6_1.to_csv('../../Data/factors/mom6m.csv')
```

# mom12m
 11-month cumulative returns ending one month before month end.

```python
mret12_1 = mret.rolling(window=11, closed='left').apply(lambda x: np.prod(x+1)-1)
mret12_1.to_csv('../../Data/factors/mom12m.csv')
```

# mom36m
Cumulative returns from months t − 36 to t − 13.

```python
mret36_12 = mret.rolling(window=24).apply(lambda x: np.prod(x+1)-1).shift(12)
mret36_12.to_csv('../../Data/factors/mom36m.csv')
```

# chmom
chmom: Cumulative returns from months t − 6 to t − 1 minus months t − 12 to t − 7.

```python
mret6_0 = mret.rolling(window=6).apply(lambda x: np.prod(x+1)-1)
mret12_6 = mret.rolling(window=6).apply(lambda x: np.prod(x+1)-1).shift(6)
chmom = mret6_0 - mret12_6
chmom.to_csv('../../Data/factors/chmom.csv')
```

# maxret
Maximum daily return from returns during month t − 1

```python
dret = mm.extract_variable(path='../../Data/raw_data/01日个股回报率',
                          fileType='csv',
                          variable='Dretwd',
                          time_col='Trddt',
                          code_col='Stkcd',
                          print_fileName=True)
dret
```

```python
maxret = dret.resample('1m').max()
maxret
```
