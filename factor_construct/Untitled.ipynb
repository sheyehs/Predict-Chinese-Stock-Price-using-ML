```python
data = mm.extract_variable(path='C:/Users/ShaunMarx/助研_人大/project01_China_stock/Data/raw_data/01日个股回报率',
                   variable='Dretwd',
                   time_col='Trddt',
                   code_col='Stkcd')
data
```

```python
window = 10
data_std = data.rolling(window=window, closed='left').apply(np.std)
data_std
```

```python
data_select = data_std.applymap(lambda x: x<0.02)
```

```python
data1 =data.mask(data_select)
data1
```

```python
target=data1.T
target
```

```python
window = 5
factors = []
for i in range(1,window+1):
    factor = data.shift(i)
    factors.append(factor.T)
```

```python
factors[1]
```

```python
date_series = target.columns
date_series
```

```python
poly = PolynomialFeatures(3)
date_tables = []
for date in date_series:
    target_column = target.loc[:,date]
    target_column.name = 'y'
    factor_columns = []
    for i in range(len(factors)):
        table = factors[i]
        column = table.loc[:,date]
        factor_columns.append(column)
    factor_table = pd.concat(factor_columns,axis=1)
    factor_table = factor_table.dropna()
    if factor_table.empty:
        continue
    factor_poly_table = pd.DataFrame(poly.fit_transform(factor_table.to_numpy()))
    date_table = pd.concat([target_column,factor_poly_table],axis=1)
    date_tables.append(date_table)
```
