# 4. chinv
Change in inventory scaled by total assets.

```python
df = pd.read_csv('../../Data/raw_data/16资产负债表/FS_Combas.csv', index_col='Accper')
df = df[df['Typrep'] == 'A']
df = df[df['Stkcd'].map(lambda x: (x>=0 and x<=9999) or (x>=600000 and x<=609999))]
df.index = pd.DatetimeIndex(df.index)
df
```

```python
total_assets = df.pivot_table(values='A001000000', index='Accper', columns='Stkcd')  # A001000000资产总计
total_assets = total_assets.replace(0,np.nan)
total_assets
```

```python
inventory = df.pivot_table(values='A001123000', index='Accper', columns='Stkcd')
# A001123000 [存货净额] - “存货”与“存货跌价准备”之差额。
inventory = inventory.replace(0,np.nan)
inventory
```

```python
inventory_by_assets = inventory / total_assets
inventory_by_assets = inventory_by_assets.replace(np.inf, np.nan)
inventory_by_assets
```

```python
pct_inventory_by_assets = inventory_by_assets.pct_change()
pct_inventory_by_assets = pct_inventory_by_assets.replace(0,np.nan)
pct_inventory_by_assets
```

```python
pct_inventory_by_assets.to_csv('../../Data/factors/chinv.csv')
```

# deprecated
outstanding shares = issued shares - treasury shares

## treasury share

```python
df = pd.read_csv('D:/csmar解压数据/16资产负债表/FS_Combas.csv', index_col='Accper')
df = df[df['Typrep'] == 'A']
f = lambda x: True if (x>=0 and x<=9999) or (x>=600000 and x<=609999) else False
df = df[df['Stkcd'].map(f)]
df.index = pd.DatetimeIndex(df.index)
df
```

```python
treasury_share = df.pivot_table(values='A003102101', index='Accper', columns='Stkcd')
# A003102101 [其中：库存股] - 企业收购、转让或注销的本公司股份金额。
treasury_share
```

```python
m_s = pd.date_range(start='1990-01-31', end='2021-12-31', freq='M')
treasury_share = treasury_share.asfreq('M').reindex(m_s).fillna(method='ffill').replace(np.nan, 0.0)
treasury_share
```

## issued share

```python
df = pd.read_csv('D:/csmar解压数据/17股本变动/TRD_Capchg.csv', index_col='Shrchgdt')
f = lambda x: True if (x>=0 and x<=9999) or (x>=600000 and x<=609999) else False
df = df[df['Stkcd'].map(f)]
df.index = pd.DatetimeIndex(df.index)
df
```

```python
issued_share = df.pivot_table(values='Nshrttl', index='Shrchgdt', columns='Stkcd')
issued_share
```

```python
m_s = pd.date_range(start='1990-01-31', end='2021-12-31', freq='M')
m_s
```

```python
new_index = issued_share.index.append(m_s).sort_values().drop_duplicates()
new_index
```

```python
issued_share = issued_share.reindex(new_index).fillna(method='ffill')
issued_share
```

```python
issued_share = issued_share.loc[m_s]
issued_share
```

```python
outstanding_share = issued_share - treasury_share
outstanding_share
```
