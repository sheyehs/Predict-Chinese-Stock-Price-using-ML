# top10holderrate
Percentage of common shares owned by top 10 shareholders

```python
# 原csv文件有问题，故改下载excel
df = pd.read_excel('../../Data/raw_data/06中国上市公司股权性质/EN_EquityNatureAll.xlsx')
df
```

```python
df = df[df['Symbol'].map(lambda x: x.startswith('00') or x.startswith('60'))]
df['EndDate'] = pd.to_datetime(df['EndDate'], format='%Y-%m-%d')
df['Symbol'] = df['Symbol'].map(lambda x: int(x))
df['TopTenHoldersRate'] = df['TopTenHoldersRate'].map(lambda x: float(x))
df
```

```python
top10holderrate = df.pivot_table(values='TopTenHoldersRate', index='EndDate', columns='Symbol')
top10holderrate
```

```python
top10holderrate.to_csv('../../Data/factors/top10holderrate.csv')
```

# largestholderrate
Percentage of common shares owned by the largest shareholder.

```python
df = pd.read_excel('../../Data/raw_data/06中国上市公司股权性质/EN_EquityNatureAll.xlsx')
df = df[df['Symbol'].map(lambda x: x.startswith('00') or x.startswith('60'))]
df
```

```python
df['EndDate'] = pd.to_datetime(df['EndDate'], format='%Y-%m-%d')
df['Symbol'] = df['Symbol'].map(lambda x: int(x))
df['LargestHolderRate'] = df['LargestHolderRate'].map(lambda x: float(x))
```

```python
largestholderrate = df.pivot_table(values='LargestHolderRate', index='EndDate', columns='Symbol')
largestholderrate
```

```python
largestholderrate.to_csv('../../Data/factors/largestholderrate.csv')
```
