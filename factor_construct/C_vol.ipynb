# idiovol
Standard deviation of residuals of weekly returns on weekly equally-weighted market
returns for three years prior to month end.

```python
df1 = pd.read_csv('../../Data/raw_data/12周个股回报率/TRD_Week.csv')
df2 = pd.read_csv('../../Data/raw_data/12周个股回报率/TRD_Week1.csv')
df3 = pd.read_csv('../../Data/raw_data/12周个股回报率/TRD_Week2.csv')
df = pd.concat([df1,df2,df3])
df = df[df['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<=609999))]
df
```

```python
df['Trdwnt'] = pd.to_datetime([x+'-5' for x in df['Trdwnt']], format='%Y-%W-%w') - pd.DateOffset(days=7)
df
```

```python
stk_wret = df.pivot_table(values='Wretwd', index='Trdwnt', columns='Stkcd')
stk_wret
```

```python
df = pd.read_csv('../../Data/raw_data/13周市场回报率/TRD_Weekcm.csv', index_col='Trdwnt')
mkt_wret = df[df['Markettype'] == 5]['Cwretwdeq']  # equally-weighted market returns
mkt_wret
```

```python
mkt_wret.index = pd.to_datetime([x+'-5' for x in mkt_wret.index], format='%Y-%W-%w') - pd.DateOffset(days=7)
mkt_wret
```

```python
month_ends = pd.date_range(start='1990-12-31', end='2021-01-31', freq='M')
month_ends
```

```python
model = LinearRegression()
idiovol = pd.DataFrame(columns=stk_wret.columns, index=month_ends)
for endday in month_ends:
    print(endday)
    beginday = endday - pd.DateOffset(years=3)
    stk_sec = stk_wret.loc[beginday:endday]
    for stock in stk_wret.columns:
        y = stk_sec.loc[:, stock]
        x = mkt_wret[beginday:endday]
        temp_df = pd.concat([y, x], axis=1).dropna()
        if len(temp_df) <=52:
            continue
        X = temp_df.iloc[:,[1]].to_numpy()
        y = temp_df.iloc[:,0].to_numpy()
        model.fit(X=X, y=y)      
        idiovol.loc[endday, stock] = np.std(y - model.predict(X))
```

```python
idiovol
```

```python
idiovol.to_csv('../../Data/factors/idiovol.csv')
```

# ear
Sum of daily returns in three days around earnings announcement.  
财报公布日附近三天的日回报率之和  
该因子如何与其他因子在日期上对齐？

```python
dret = mm.extract_variable(path='../../Data/raw_data/01日个股回报率',
                           fileType='csv',
                           variable='Dretwd',
                           time_col='Trddt',
                           code_col='Stkcd',
                           print_fileName=True
                      )
dret
```

```python
df = pd.read_csv('../../Data/raw_data/18年、中、季报基本情况/IAR_Rept.csv')
df
```

```python
df = df[df['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<= 609999))]
df
```

```python
df['Annodt'] = pd.to_datetime(df['Annodt'], format='%Y-%m-%d')
```

```python
stk_annodt = df[['Stkcd','Annodt']]
stk_annodt
```

```python
stk_annodt = stk_annodt.sort_values(by='Annodt')
stk_annodt
```

```python
dret_s3d = dret.rolling(window=3, center=True).sum()
dret_s3d
```

```python
ear = pd.DataFrame(columns=dret_s3d.columns, index=pd.date_range(start='1990-01-01', end='2022-02-28'))
for i in range(len(stk_annodt)):
    stk = stk_annodt.iloc[i,0]
    adt = stk_annodt.iloc[i,1]
    if stk in dret_s3d.columns:
        if adt in dret_s3d.index:  # 如果公布日是交易日
            ear.loc[adt, stk] = dret_s3d.loc[adt, stk]
        else:  # 如果公布日不是交易日
            afterdates = dret_s3d.loc[adt:, stk]
            if not afterdates.empty:
                ear.loc[adt, stk] = afterdates[0]
```

```python
ear
```

```python
ear.to_csv('../../Data/factors/ear.csv')
```

# roavol
 Standard deviation of 16 quarters of income before extraordinary items divided by
average total assets.

```python
# B001100000 [营业总收入] - 企业经营过程中所有收入之和.
df = pd.read_csv('../../Data/raw_data/14利润表/FS_Comins.csv')
df = df[df['Typrep'] == 'A' ]
df = df[df['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<=609999))]
df['Accper'] = pd.to_datetime(df['Accper'], format='%Y-%m-%d')
df
```

```python
income = df.pivot_table(values='B001100000', index='Accper', columns='Stkcd')
#income = income.fillna(method='ffill')
income
```

```python
# A001000000 [资产总计] - 资产各项目之总计.
df = pd.read_csv('../../Data/raw_data/16资产负债表/FS_Combas.csv')
df = df[df['Typrep'] == 'A' ]
df = df[df['Stkcd'].map(lambda x: (0<=x and x<=9999) or (600000<=x and x<=609999))]
df['Accper'] = pd.to_datetime(df['Accper'], format='%Y-%m-%d')
df
```

```python
assets = df.pivot_table(values='A001000000', index='Accper', columns='Stkcd')
assets
```

```python
aver_assets = assets.rolling(f'{16*3*30}D').mean()
aver_assets
```

```python
roavol = income.rolling(window=f'{16*3*30}D',min_periods=2).std() / aver_assets  # std([x1/yn,...,xn/yn]) = std([x1,...,xn]) / yn
roavol
```

```python
roavol.to_csv('../../Data/factors/roavol.csv')
```
