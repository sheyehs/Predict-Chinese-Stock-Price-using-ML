```python
import my_methods as mm
import pandas as pd
import numpy as np
```

```python
stkcls = pd.read_csv('../../Data/raw_data/05上市公司行业分类表/STK_INDUSTRYCLASS.csv')
stkcls = stkcls[stkcls['IndustryClassificationID'] == 'P0207']
stkcls = stkcls[['Symbol','IndustryCode','ImplementDate']]
stkcls = stkcls[stkcls['Symbol'].map(lambda x: (x>=0 and x<=9999) or (x>=600000 and x<=609999))]
stkcls['ImplementDate'] = pd.DatetimeIndex(stkcls['ImplementDate'])
stkcls
```

```python
stkcls['IndustryCode'].value_counts()
```

```python
df = stkcls.pivot_table(values='IndustryCode',index='ImplementDate',columns='Symbol',aggfunc=lambda x:x)
df = df.fillna(value='None')  # 不都改成文本，run_model那边总会报数据类型混杂warning
df
```
